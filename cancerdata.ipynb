```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
label=le.fit_transform(data['diagnosis'])
```

```python
data.drop("diagnosis", axis=1, inplace=True)
data["diagnosis"] = label
```

```python
data.head()
```

```python
#the column unamed is useless as there is no values in it

newdata=data.drop(['id','Unnamed: 32'],axis=1)
```

```python
sns.set(rc = {'figure.figsize':(30,15)})
sns.heatmap(newdata.corr(),annot=True)
```

```python
# importing the required libraries
import matplotlib.pyplot as plt
import numpy as np

# define data values
x = newdata.drop('diagnosis',axis='columns') # X-axis points
y = newdata.diagnosis # Y-axis points

plt.plot(x, y) # Plot the chart
plt.show() # display
```

```python
sns.lineplot(data=newdata)
```

```python
x.columns
```

# training the dataset

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
X_train
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf=RandomForestClassifier(n_estimators = 100, random_state = 0)
```

```python
clf.fit(X_train,y_train)
```

```python
ypred=clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ypred,y_test)
```

```python
import pickle
pickle.dump(clf,open('model.pkl','wb'))
```
